```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 equidistant points between -5 to 5 for x
x = np.linspace(-5, 5, 100)

# Calculate y values for each equation
y1 = (3 - 2*x) / 3
y2 = (6 - 4*x) / 6
y3 = (1 - 3*x) / (-2)

# Generate a figure frame for plotting the lines
fig, ax = plt.subplots()

# Plot the three lines
ax.plot(x, y1, color='r', label='$2x + 3y = 3$')
ax.plot(x, y2, color='g', label='$4x + 6y = 6$')
ax.plot(x, y3, color='b', label='$3x - 2y = 1$')

# Set legend
ax.legend()

# Calculate the intersection point of the three lines
intersection_x = 0.2  # Just an initial guess for x value
intersection_y = (3 - 2*intersection_x) / 3

# Plot the point of intersection
ax.plot(intersection_x, intersection_y, 'bo')  # Blue dot for intersection point

# Annotate the intersection point
ax.text(intersection_x, intersection_y, f'({intersection_x:.2f}, {intersection_y:.2f})',
        verticalalignment='bottom', horizontalalignment='right')

# Set the limits of the X & Y axes
ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)

# Set property of X and Y axes
ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)
ax.set_xticks(np.arange(-2, 4, 1))
ax.set_yticks(np.arange(-2, 4, 1))

# Set aspect ratio and grid
ax.set_aspect('equal')
ax.grid(True, ls=':')

# Label axes
plt.xlabel("X")
plt.ylabel("Y")

plt.title("Linear System b")

# Show the plot
plt.show()
```

```python
import numpy as np

# Task 2: Program 1 - Adding Two Matrices
print("Task 2: Program 1 - Adding Two Matrices")
matrix1 = np.random.rand(5, 6)
matrix2 = np.random.rand(5, 6)
result_matrix = matrix1 + matrix2
print("Matrix 1:")
print(matrix1)
print("\nMatrix 2:")
print(matrix2)
print("\nResult Matrix:")
print(result_matrix)
print("\n")

# Task 2: Program 2 - Subtracting and Adding Matrices
print("Task 2: Program 2 - Subtracting and Adding Matrices")
matrix1 = np.random.rand(2, 3)
matrix2 = np.random.rand(2, 3)
matrix3 = np.random.rand(2, 3)
result_matrix = matrix1 - matrix2 + matrix3
print("Matrix 1:")
print(matrix1)
print("\nMatrix 2:")
print(matrix2)
print("\nMatrix 3:")
print(matrix3)
print("\nResult Matrix:")
print(result_matrix)
print("\n")

# Task 2: Program 3 - Finding a Number in a Matrix
print("Task 2: Program 3 - Finding a Number in a Matrix")
matrix = np.random.randint(1, 21, (4, 4))
while True:
    try:
        user_input = int(input("Enter a number to search in the matrix: "))
        break
    except ValueError:
        print("Please enter a valid number.")

if user_input in matrix:
    count = np.count_nonzero(matrix == user_input)
    print(f"The number {user_input} is found in the matrix {count} times.")
else:
    print("The number is not in the matrix.")
print("\n")

# Explanation of Numpy linalg functions
print("Explanation of Numpy linalg functions:")
a = np.array([[1, 2], [3, 4]])

# linalg.det(a): calculates the determinant of the matrix
det_a = np.linalg.det(a)
print("Determinant of matrix a:")
print(det_a)

# linalg.matrix_rank(a): computes the rank of the matrix
rank_a = np.linalg.matrix_rank(a)
print("\nRank of matrix a:")
print(rank_a)

# linalg.inv(a): calculates the multiplicative inverse of a matrix
a_inv = np.linalg.inv(a)
print("\nInverse of matrix a:")
print(a_inv)

# linalg.solve(a, b): solve a linear system of equations ax=b
b = np.array([1, 2])
x = np.linalg.solve(a, b)
print("\nSolution of linear system:")
print(x)

# linalg.pinv(a): compute the pseudo-inverse of a matrix.
a = np.random.randn(3, 2)
b = np.linalg.pinv(a)
print("\nPseudo-inverse of matrix a:")
print(b)
print("Is the product of a and its pseudo-inverse close to the identity matrix?")
print(np.allclose(a, np.dot(a, np.dot(b, a))))
```

```python
import numpy as np

# Task 3: Part 1 - Calculate the rank of a random matrix and check if it's invertible
print("Task 3: Part 1 - Calculate the rank of a random matrix and check if it's invertible")

# Create a random matrix A of order 5x5
A = np.random.rand(5, 5)

# Calculate the rank of matrix A
rank_A = np.linalg.matrix_rank(A)
print("Rank of matrix A:", rank_A)

# Check if the matrix A is invertible (i.e., non-singular)
if rank_A == min(A.shape):
    print("Matrix A is invertible (non-singular)")
else:
    print("Matrix A is not invertible (singular)")

# Other ways to check if a matrix is invertible:
# 1. Determinant: If the determinant of A is non-zero, then A is invertible.
det_A = np.linalg.det(A)
if det_A != 0:
    print("Matrix A is invertible based on the determinant check")
else:
    print("Matrix A is not invertible based on the determinant check")

# 2. Checking if the matrix A has full rank (i.e., the number of linearly independent columns equals its order).
if rank_A == min(A.shape):
    print("Matrix A has full rank, hence it is invertible")
else:
    print("Matrix A does not have full rank, hence it is not invertible")

print("\n")

# Task 3: Part 2 - Solve linear system Ax = b
print("Task 3: Part 2 - Solve linear system Ax = b")

# Generate a random matrix A of order 5x5 and a random vector x of length 5
A = np.random.rand(5, 5)
x = np.random.rand(5, 1)

# Calculate the vector b = Ax
b = np.dot(A, x)

# Solve the linear system Ax = b
x_solution = np.linalg.solve(A, b)
print("Solution for x:")
print(x_solution)

print("\n")

# Task 3: Part 3 - Solve a specific linear system
print("Task 3: Part 3 - Solve a specific linear system")

# Coefficients matrix A and constant vector b for the given system of equations
A = np.array([[8, 3, -2], [-4, 7, 5], [3, 4, -12]])
b = np.array([9, 15, 35])

# Solve the linear system Ax = b
x_solution = np.linalg.solve(A, b)
print("Solution for x, y, z:")
print(x_solution)
```
